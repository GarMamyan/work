```python
from pydantic_settings import BaseSettings
from datetime import datetime, timedelta
import requests
import json

class Settings(BaseSettings):
    """
    Settings for the application.
    """

    lead_list_url: str = (
        "https://{company_id}.bitrix24.ru/rest/{user_id}/{webhook}/crm.lead.list.json"
    )
    user_list_url: str = 'https://{company_id}.bitrix24.ru/rest/{user_id}/{webhook}/user.get.json'


settings = Settings()
```

```python
def get_leads_list_schema(date_from: str, date_to: str, offset=0):

    payload = {
        "select": ["*", "UF_*"],
        "start": offset,
        "filter": {
            ">DATE_CREATE": f"{date_from}T00:00:00",
            "<DATE_CREATE": f"{date_to}T23:59:59",
        },
        "order": {"STATUS_ID": "ASC"},
    }

    return payload
```
